```python
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual
from ipywidgets import Layout, Button, Box, FloatText, Text, Dropdown, Label,BoundedFloatText


display(HTML('<h1 style="text-align: center;">AEPsych Visualization tool</h1>'))

style = {'description_width': 'initial', "padding": "10px"}


# File uploader 
uploader = FileUpload(
    description="Resume Session",
    accept=".db",
    multiple=False
    )


# Outcome Labels O and 1
# outcome_label = Label(value='Outcome Labels:')
# display(outcome_label)

style = {'description_width': 'initial','border': '2px solid red',}


inputZero = Dropdown(
    options=[('Yes Trial', "Detected Trial"), ('No Trial', "Undetected Trial")],
    value="Detected Trial",
    description='0:',
    style=style
)
inputOne = Dropdown(
    options=[ ('No Trial', "Undetected Trial"), ('Yes Trial', "Detected Trial")],
    value="Undetected Trial",
    description='1:',
    style=style
)

# Float inputs 
target_level = BoundedFloatText(
    value=0.75,
    min=0,
    max=1,
    step=0.1,
    description='target_level:',
    disabled=False,
    style=style
)

cred_level = BoundedFloatText(
    value=0.95,
    min=0,
    max=1,
    step=0.1,
    description='cred_level:',
    disabled=False,
    style=style
)


# X and Y axis labels 

xlabel = Text(
    value='Angle (degrees)',
    placeholder='x axis',
    description='xlabel:',
    disabled=False
)

ylabel = Text(
    value='Detection Probability',
    placeholder='y axis',
    description='ylabel:',
    disabled=False
)

#  Upload db file 

# def upload_db_file(uploader):
#     return uploader

# db_file = interactive(upload_db_file, uploader=uploader)
# display(db_file)
# db_file.result
# db_file.kwargs 

# # Update inputs 
def update_plot(
    uploader,
    inputZero,
    inputOne,
    target_level,
    cred_level,
    xlabel,
    ylabel
):
    
    if uploader:
        print("File uploaded successfully! Resuming session...")
    if uploader == {}: 
        print("Starting new session...") 
#     print(inputZero,
#           inputOne, 
#           target_level,
#           cred_level,
#           xlabel,
#           ylabel)
#      you would return the plot 
    return plot_strat(strat, xlabel=xlabel, ylabel=ylabel, yes_label=inputOne, no_label=inputZero, cred_level=cred_level, target_level=target_level)



submit = interactive(update_plot,
                     {'manual': True},
                     uploader=uploader,
                     inputZero=inputZero, 
                     inputOne=inputOne,
                     target_level=target_level,
                     cred_level=cred_level,
                     xlabel=xlabel,
                     ylabel=ylabel);

submit.children[7].description = "Submit"
submit


form_item_layout = Layout(
#     display="flex",
#     flex_flow ="row",
#     justify_content="space-between",
#     margin="10px",
#     justify_content="center"
)


form_items = [
#     Box([Label(value=None), uploader], layout=form_item_layout,),
#     Box([Label(value="Outcome labels: ")], layout=label_style),
#     Box([Label(value=None), inputZero, inputOne], layout=form_item_layout),
#     Box([Label(value=None), target_level, cred_level], layout=form_item_layout),
#     Box([Label(value=None), xlabel, ylabel], layout=form_item_layout),
    Box([Label(value=None),submit],layout=form_item_layout)
]

form = Box(form_items, layout=Layout(
    display='flex',
    flex_flow='column',
#     border='solid 1px',
    align_items='stretch',
    width='100%',
    padding="20px",
))
form
```
